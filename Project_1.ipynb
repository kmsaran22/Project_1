```python
with open(r"C:\Users\lenovo\Desktop\DS\Internshala Materials\Project_1\C1-Project-File Handling & String Manipulation-Dataset\conv.txt","r+") as txt_file:
    txt = txt_file.read()
txt 
```

```python
txt=txt.upper()
txt
```

```python
txt = txt.replace('WAYMAR ROYCE','ROYCE')
punc = ',.?!'
txt1=""
for char in txt:
    if char not in punc:
        txt1=txt1+char
txt1        
```

```python
sample1 = txt1.split("\n\n")
sample1.remove('')
```

```python
sample1
```

```python
sample2=[]
for i in range (len(sample1)):
    sample2.append (sample1[i].split(':'))
sample2  
```

```python
sample3=[]
for i in range (len(sample2)):
    sample3.append(sample2[i][0])
sample3 = set(sample3)
sample3 = list(sample3)
sample3    # gives the name of unique dialoge speakers
```

```python
# number of unique dialoge speakers
len(sample3)
```

```python
sample5=[]
for i in sample3:
    sample4=[]
    for j in sample2:
        if (i==j[0]):
            sample4.extend((j[1]).split(' '))
            sample4=set(sample4)
            sample4=list(sample4)
            sample4.sort()
            sample4.remove('')
    sample5.append([i,sample4])
    
sample5 
```

```python
#to  find the number of unique words
for i in sample5:
    
    for j in i[1]:
        count = 0
        for k in sample5:
            if (i[0]==k[0]):
                continue
            else:
                if j in k[1]:
                    k[1].remove(j)
                    count = count+1
        if(count!=0):
            i[1].remove(j)
```

```python
#to write unique words to correponding file

for i in sample5:
    path = 'C:/Users/lenovo/Desktop/DS/Internshala Learning/Project_sample/' 
    file_name = str(i[0])+".txt"
    with open(path+file_name,'w+') as f:
        for j in i[1]:
            f.write(j+"\n")
```

```python
#to find the total number of unique words

sample6=[]
for i in sample5:
    for j in i[1]:
        sample6.append(j)
len(sample6) 
```
